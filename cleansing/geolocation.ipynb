```python
import pandas as pd
```

```python
df = pd.read_csv('/content/jog.csv')
```

```python
from geopy.geocoders import Nominatim
```

```python
df['lokasi'] = df['lokasi'].str.split(',').str[0]
```

```python
geolocator = Nominatim(user_agent="my_geocoder",timeout=10)

# Fungsi untuk mendapatkan koordinat dari kecamatan di kota Yogyakarta
def get_coordinates(lokasi):
    viewbox = [110.3600, -7.9300, 110.4600, -7.7900]
    full_address = f"{lokasi}, Yogyakarta, Indonesia"
    location = geolocator.geocode(full_address)

    if location:
        return location.latitude, location.longitude
    else:
        return None, None
df[['latitude', 'longitude']] = df['lokasi'].apply(get_coordinates).apply(pd.Series)
df.head()
```

```python
df.to_csv('jogja_.csv')
```
