```python
import pandas as pd
```

```python
df = pd.read_csv('/content/jogja___.csv')
```

```python
fitur_1 = [
    'sistem_alarm','gym','internet_broadband_wifi','tv_kabel','pemanas_ruangan','pendingin_ruangan',
    'air_panas','telepon','televisi','kitchen_set','garasi','secure_parking','kolam_renang','lapangan_tenis',
    'balkon','dek','halaman_terbuka','area_hiburan_outdoor','pagar_penuh','taman','keamanan_24_jam','taman_bermain_anak'
]
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Menentukan jumlah subplot sesuai dengan jumlah kolom yang akan divisualisasikan
num_plots = len(fitur_1)
num_rows = int(np.ceil(np.sqrt(num_plots)))
num_cols = int(np.ceil(num_plots / num_rows))

# Membuat subplot
fig, axes = plt.subplots(num_rows, num_cols, figsize=(12, 10))
fig.suptitle('fitur')

# Flatten array dari subplot untuk mempermudah iterasi
axes = axes.flatten()

# Iterasi untuk setiap kolom dan membuat pie chart
for i, column in enumerate(fitur_1):
    ax = axes[i]
    category_counts = df[column].value_counts()

    ax.pie(category_counts,
           labels=category_counts.index,
           autopct='%1.1f%%',
           startangle=140,
           labeldistance=0.3)

    ax.set_title(f'{column}')
    ax.axis('equal')

# Menyembunyikan subplot yang tidak digunakan
for j in range(i+1, len(axes)):
    axes[j].axis('off')

plt.tight_layout()
plt.show()
```

```python
total_fitur = len(fitur_2)
subplot_rows = int(np.ceil(total_fitur / 3))
subplot_cols = min(3, total_fitur)

fig, axes = plt.subplots(subplot_rows, subplot_cols, figsize=(15, 2 * subplot_rows))
fig.suptitle('Lokasi', y=1.02)

# Flatten axes jika lebih dari satu baris
axes = axes.flatten()

for i, ax in enumerate(axes):
    if i < total_fitur:
        column = fitur_2[i]
        category_counts = df[column].value_counts()

        ax.bar(category_counts.index,
               category_counts.values,
               tick_label=category_counts.index,
               color='skyblue')  # Ganti warna sesuai keinginan

        ax.set_title(f'{column}')
plt.tight_layout()
plt.show()
```
